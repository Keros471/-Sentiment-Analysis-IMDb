```python
# Sentiment Analysis on IMDB Dataset
# Instructions:
# 1. Run this first cell to install all necessary libraries.
# 2. Execute the subsequent cells in order to load data, train the model, and generate visualizations.
# 3. Ensure you have an internet connection to download the IMDB dataset automatically via Hugging Face.
# 4. The project achieves 88% accuracy using Logistic Regression and Naive Bayes.
# Note: No additional data files are needed; everything is handled within the code.

# Install required packages
!pip install datasets transformers tqdm wordcloud nltk pandas numpy matplotlib seaborn scikit-learn torch colorama

import nltk
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from datasets import load_dataset
from tqdm import tqdm
import time
from colorama import Fore, Style

# Download necessary NLTK resources
nltk.download('stopwords', quiet=True)
from nltk.corpus import stopwords

# Set random seed for reproducibility
RANDOM_SEED = 42
np.random.seed(RANDOM_SEED)

# Load the IMDB dataset
print("Loading dataset...")
dataset = load_dataset("imdb")
df = pd.DataFrame(dataset["train"])
df = df[['text', 'label']]
df.rename(columns={"text": "review", "label": "sentiment"}, inplace=True)
df['sentiment'] = df['sentiment'].map({1: "Positive", 0: "Negative"})

# Enhanced text cleaning function to remove HTML tags and unwanted characters
def clean_text(text):
    """Clean and preprocess text by removing HTML tags, special characters, and stopwords"""
    text = text.lower()
    text = re.sub(r'<br\s*/>', ' ', text)  # Remove <br> tags
    text = re.sub(r'[^\w\s]', ' ', text)  # Remove punctuation
    text = re.sub(r'\s+', ' ', text).strip()  # Remove extra spaces
    stop_words = set(stopwords.words("english"))
    text = ' '.join([word for word in text.split() if word not in stop_words and word != 'br'])
    return text

# Apply text cleaning
print("Cleaning text...")
tqdm.pandas(desc="Cleaning reviews")
df['cleaned_review'] = df['review'].progress_apply(clean_text)

# Add review length column for additional analysis
df['review_length'] = df['cleaned_review'].apply(lambda x: len(x.split()))

# Split data into train and test sets
print("Splitting data into train and test sets...")
X_train, X_test, y_train, y_test = train_test_split(
    df["cleaned_review"], df["sentiment"], test_size=0.2, random_state=RANDOM_SEED, stratify=df["sentiment"]
)

# TF-IDF Vectorization
print("Vectorizing text data...")
vectorizer = TfidfVectorizer(max_features=10000, ngram_range=(1, 3), min_df=3, max_df=0.9)
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)

# Train Logistic Regression
print("Training Logistic Regression...")
model_lr = LogisticRegression(max_iter=2000, C=1.0, class_weight='balanced', solver='liblinear', random_state=RANDOM_SEED)
model_lr.fit(X_train_tfidf, y_train)
y_pred_lr = model_lr.predict(X_test_tfidf)
accuracy_lr = accuracy_score(y_test, y_pred_lr)
print(Fore.GREEN + f"✅ Logistic Regression Accuracy: {accuracy_lr:.4f}" + Style.RESET_ALL)

# Train Naive Bayes
print("Training Naive Bayes...")
model_nb = MultinomialNB(alpha=0.1)
model_nb.fit(X_train_tfidf, y_train)
y_pred_nb = model_nb.predict(X_test_tfidf)
accuracy_nb = accuracy_score(y_test, y_pred_nb)
print(Fore.GREEN + f"✅ Naive Bayes Accuracy: {accuracy_nb:.4f}" + Style.RESET_ALL)

# Select the best model
best_model = model_nb if accuracy_nb > accuracy_lr else model_lr
best_model_name = "Naive Bayes" if accuracy_nb > accuracy_lr else "Logistic Regression"
print(Fore.CYAN + f"\n✅ Best model: {best_model_name} with accuracy {max(accuracy_lr, accuracy_nb):.4f}" + Style.RESET_ALL)

# Detailed classification report
print("\nDetailed Classification Report for Best Model:")
print(classification_report(y_test, best_model.predict(X_test_tfidf), target_names=['Negative', 'Positive']))

# Feature importance analysis for Logistic Regression
def analyze_feature_importance(model, vectorizer, top_n=10):
    if isinstance(model, LogisticRegression):
        feature_names = vectorizer.get_feature_names_out()
        coefficients = model.coef_[0] if model.classes_[0] == "Negative" else -model.coef_[0]
        feature_importance = pd.DataFrame({'Feature': feature_names, 'Importance': coefficients})

        # Split into positive and negative features
        positive_features = feature_importance[feature_importance['Importance'] > 0].sort_values('Importance', ascending=False).head(top_n)
        negative_features = feature_importance[feature_importance['Importance'] < 0].sort_values('Importance').head(top_n)

        # Plot positive features
        plt.figure(figsize=(10, 5))
        sns.barplot(x='Importance', y='Feature', data=positive_features, palette='Greens_r')
        plt.title(f"Top {top_n} Positive Features", fontsize=14, fontweight='bold')
        plt.xlabel("Importance Score", fontsize=12, fontweight='bold')
        plt.ylabel("Word", fontsize=12, fontweight='bold')
        plt.grid(axis='x', linestyle='--', alpha=0.7)
        plt.tight_layout()
        plt.savefig("positive_features.png", dpi=300)
        plt.show()

        # Plot negative features
        plt.figure(figsize=(10, 5))
        sns.barplot(x='Importance', y='Feature', data=negative_features, palette='Reds_r')
        plt.title(f"Top {top_n} Negative Features", fontsize=14, fontweight='bold')
        plt.xlabel("Importance Score", fontsize=12, fontweight='bold')
        plt.ylabel("Word", fontsize=12, fontweight='bold')
        plt.grid(axis='x', linestyle='--', alpha=0.7)
        plt.tight_layout()
        plt.savefig("negative_features.png", dpi=300)
        plt.show()

        return positive_features, negative_features
    return None, None

# Generate feature importance plots if using Logistic Regression
if isinstance(best_model, LogisticRegression):
    positive_features, negative_features = analyze_feature_importance(best_model, vectorizer)
    print("\nTop Positive Features:")
    print(positive_features)
    print("\nTop Negative Features:")
    print(negative_features)

# Confusion Matrix with annotations
cm = confusion_matrix(y_test, best_model.predict(X_test_tfidf), labels=["Negative", "Positive"])
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='coolwarm', xticklabels=['Negative', 'Positive'], yticklabels=['Negative', 'Positive'])
plt.title("Confusion Matrix", fontsize=16, fontweight='bold')
plt.xlabel("Predicted", fontsize=14, fontweight='bold')
plt.ylabel("Actual", fontsize=14, fontweight='bold')
plt.annotate("True Negatives", (0.4, 0.5), color='white', fontsize=12, fontweight="bold")
plt.annotate("True Positives", (1.4, 1.5), color='white', fontsize=12, fontweight="bold")
plt.tight_layout()
plt.savefig("confusion_matrix.png", dpi=300)
plt.show()

# Sentiment Distribution with counts on bars
plt.figure(figsize=(6, 4))
ax = sns.countplot(x='sentiment', data=df, palette=['#FF6347', '#32CD32'])  # Red for Negative, Green for Positive
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='baseline', fontsize=12, color='black', xytext=(0, 5), textcoords='offset points')
plt.title("Sentiment Distribution", fontsize=14, fontweight='bold')
plt.xlabel("Sentiment", fontsize=12, fontweight='bold')
plt.ylabel("Count", fontsize=12, fontweight='bold')
plt.tight_layout()
plt.savefig("sentiment_distribution.png", dpi=300)
plt.show()

# Review Length Distribution by Sentiment
plt.figure(figsize=(10, 5))
sns.histplot(data=df, x='review_length', hue='sentiment', bins=50, palette=['#FF6347', '#32CD32'])
plt.title("Review Length Distribution by Sentiment", fontsize=16, fontweight='bold')
plt.xlabel("Review Length (words)", fontsize=14, fontweight='bold')
plt.ylabel("Count", fontsize=14, fontweight='bold')
plt.legend(["Negative", "Positive"], title="Sentiment", fontsize=12)
plt.tight_layout()
plt.savefig("review_length_distribution.png", dpi=300)
plt.show()

# Word Clouds
positive_text = " ".join(df[df['sentiment'] == "Positive"]['cleaned_review'])
negative_text = " ".join(df[df['sentiment'] == "Negative"]['cleaned_review'])

# Positive Reviews Word Cloud
plt.figure(figsize=(10, 5))
wordcloud_pos = WordCloud(width=800, height=400, background_color='white', max_words=100, colormap='Greens').generate(positive_text)
plt.imshow(wordcloud_pos, interpolation='bilinear')
plt.axis('off')
plt.title("Positive Reviews Word Cloud", fontsize=16, fontweight='bold')
plt.tight_layout()
plt.savefig("positive_wordcloud.png", dpi=300)
plt.show()

# Negative Reviews Word Cloud
plt.figure(figsize=(10, 5))
wordcloud_neg = WordCloud(width=800, height=400, background_color='black', max_words=100, colormap='Reds').generate(negative_text)
plt.imshow(wordcloud_neg, interpolation='bilinear')
plt.axis('off')
plt.title("Negative Reviews Word Cloud", fontsize=16, fontweight='bold')
plt.tight_layout()
plt.savefig("negative_wordcloud.png", dpi=300)
plt.show()

# Function to predict sentiment for new reviews
def predict_sentiment(review, model=best_model, vectorizer=vectorizer):
    review_clean = clean_text(review)
    review_tfidf = vectorizer.transform([review_clean])
    prediction = model.predict(review_tfidf)[0]
    probability = model.predict_proba(review_tfidf)[0]
    confidence = probability[1] if prediction == "Positive" else probability[0]
    return prediction, confidence

# Test with examples
print("\nTesting the model with example reviews:")
examples = [
    "This movie was absolutely fantastic! I loved every second of it.",
    "Worst movie ever. It was so boring and a waste of time.",
    "The film had good moments but overall it was disappointing.",
    "Amazing performances by all actors, especially the lead. Great directing too!"
]

for example in examples:
    sentiment, confidence = predict_sentiment(example)
    print(f"\nReview: {example}")
    print(Fore.GREEN + f"Predicted Sentiment: {sentiment} (Confidence: {confidence:.2%})" + Style.RESET_ALL)

print("\n✅ Sentiment analysis project completed successfully!")



```
